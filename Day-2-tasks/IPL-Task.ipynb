```python
#importing datasets

import pandas as pd

match_path='/Users/user/Desktop/python projects/Match_csv.csv'
match_df=pd.read_csv(match_path)
player_path='/Users/user/Desktop/python projects/player_csv.csv'
player_df=pd.read_csv(player_path)
team_path='/Users/user/Desktop/python projects/Team_csv.csv'
team_df=pd.read_csv(team_path)
runs_path='/Users/user/Desktop/python projects/runs.csv'
runs_df=pd.read_csv(runs_path)
print(match_df)
print(player_df)
print(team_df)
print(runs_df)
```

```python
#Task-1

# Map team names from team_df to match_df
team_dict = team_df.set_index('Team_Id')['Team_Name'].to_dict()

# Add team names to the match dataframe
match_df['Team_Name'] = match_df['Team_Name_Id'].map(team_dict)
match_df['Opponent_Team_Name'] = match_df['Opponent_Team_Id'].map(team_dict)

# Calculate total matches played by each team
matches_played = match_df['Team_Name'].value_counts().reset_index()
matches_played.columns = ['Team_Name', 'Total_Matches_Played']

# Calculate total matches won by each team
matches_won = match_df['Match_Winner_Id'].map(team_dict).value_counts().reset_index()
matches_won.columns = ['Team_Name', 'Total_Matches_Won']

# Merge the two dataframes to create the final dataframe
team_stats_df = pd.merge(matches_played, matches_won, on='Team_Name')

# Display the resulting dataframe
print(team_stats_df)
```

```python
#Task-2 

print("player_df columns:", player_df.columns)
print("runs_df columns:", runs_df.columns)

# Merge runs_df with player_df using Striker_Id from runs_df and Player_Id from player_df
runs_with_names = pd.merge(runs_df, player_df[['Player_Id', 'Player_Name']], left_on='Striker_Id', right_on='Player_Id')

# Display the top 10 run scorers in the IPL
# Group by Player_Name and sum the runs scored
top_scorers = runs_with_names.groupby('Player_Name')['Batsman_Scored'].sum().reset_index()

# Sort by runs scored in descending order and get the top 10
top_scorers = top_scorers.sort_values(by='Batsman_Scored', ascending=False).head(10)

# Display the top 10 run scorers
print("Top 10 Run Scorers in the IPL:")
print(top_scorers)

# Display the top 10 batting performances in terms of runs scored in a match
# Sort by runs scored in descending order to find the top 10 individual match performances
top_performances = runs_with_names.sort_values(by='Batsman_Scored', ascending=False).head(10)

# Display the top 10 batting performances in a match
print("Top 10 Batting Performances in a Match:")
print(top_performances[['Player_Name', 'Match_Id', 'Batsman_Scored']])
```

```python
#Task-3

print(runs_df.head())



# Merge runs_df with player_df using Striker_Id from runs_df and Player_Id from player_df
runs_with_names = pd.merge(runs_df, player_df[['Player_Id', 'Player_Name']], left_on='Striker_Id', right_on='Player_Id')

# Group by Player_Name to calculate total runs and total balls faced
player_stats = runs_with_names.groupby('Player_Name').agg({
    'Batsman_Scored': 'sum',
    'Ball_Id': 'sum'
}).reset_index()

# Calculate the strike rate
player_stats['Strike_Rate'] = (player_stats['Batsman_Scored'] / player_stats['Ball_Id']) * 100

# Sort by strike rate in descending order and get the top 10 players
top_strike_rates = player_stats.sort_values(by='Strike_Rate', ascending=False).head(10)

# Display the top 10 players by strike rate
print("Top 10 Players by Strike Rate:")
print(top_strike_rates[['Player_Name', 'Batsman_Scored', 'Ball_Id', 'Strike_Rate']])
```

```python
#Task-4



# Merge runs_df with player_df using Bowler_Id from runs_df and Player_Id from player_df
bowling_with_names = pd.merge(runs_df, player_df[['Player_Id', 'Player_Name']], left_on='Bowler_Id', right_on='Player_Id')

# Top 10 players in terms of wickets taken
# Group by Player_Name to calculate total wickets taken
wicket_takers = bowling_with_names.groupby('Player_Name')['Player_dissimal_Id'].sum().reset_index()

# Sort by wickets taken in descending order and get the top 10 players
top_wicket_takers = wicket_takers.sort_values(by='Player_dissimal_Id', ascending=False).head(10)

# Display the top 10 players by wickets taken
print("Top 10 Players by Wickets Taken:")
print(top_wicket_takers)

#  Top 10 players in terms of economic rate
# Group by Player_Name to calculate total runs conceded and total balls bowled
bowling_stats = bowling_with_names.groupby('Player_Name').agg({
    'Batsman_Scored': 'sum',
    'Ball_Id': 'sum'
}).reset_index()

# Calculate the economic rate
bowling_stats['Overs_Bowled'] = bowling_stats['Ball_Id'] / 6
bowling_stats['Economy_Rate'] = bowling_stats['Batsman_Scored'].sum() / bowling_stats['Overs_Bowled'].sum()

# Sort by economic rate in ascending order (lower is better) and get the top 10 players
top_economy_rate = bowling_stats.sort_values(by='Economy_Rate', ascending=True).head(10)

# Display the top 10 players by economic rate
print("Top 10 Players by Economic Rate:")
print(top_economy_rate[['Player_Name', 'Batsman_Scored', 'Ball_Id', 'Economy_Rate']])
```

```python
#Task-6


# Check if 'CSK' is listed in teams data
csk_team_id = team_df[team_df['Team_Name'] == 'Chennai Super Kings']['Team_Id'].values[0]

# Filter matches where CSK played
csk_matches = match_df[(match_df['Team_Name_Id'] == csk_team_id) | (match_df['Opponent_Team_Id'] == csk_team_id)]

# Determine whether CSK batted first or second
csk_matches['Toss_Winner_Id'] = csk_matches.apply(
    lambda row: row['Team_Name_Id'] == csk_team_id if row['Team_Name_Id'] == row['Toss_Winner_Id'] else row['Opponent_Team_Id'] == csk_team_id,
    axis=1
)

# Separate matches into those where CSK batted first and second
csk_bat_first = csk_matches[csk_matches['Toss_Decision'] == 'bat']
csk_bat_second = csk_matches[csk_matches['Toss_Decision'] == 'field']

# Calculate win percentage when CSK bats first
csk_win_first = csk_bat_first[csk_bat_first['Match_Winner_Id'] == csk_team_id]
csk_win_percentage_first = len(csk_win_first) / len(csk_bat_first) * 100 if len(csk_bat_first) > 0 else 0

# Calculate win percentage when CSK bats second
csk_win_second = csk_bat_second[csk_bat_second['Match_Winner_Id'] == csk_team_id]
csk_win_percentage_second = len(csk_win_second) / len(csk_bat_second) * 100 if len(csk_bat_second) > 0 else 0

# Output the win percentages
print(f"CSK Win Percentage when batting first: {csk_win_percentage_first:.2f}%")
print(f"CSK Win Percentage when batting second: {csk_win_percentage_second:.2f}%")

# Determine if CSK is better when batting second
if csk_win_percentage_second > csk_win_percentage_first:
    print("CSK is a better team when they bat second.")
else:
    print("CSK is not necessarily a better team when they bat second.")
```

```python
#Task-7 

# Check if 'Virat Kohli' is listed in players data
kohli_player_id = player_df[player_df['Player_Name'] == 'V Kohli']['Player_Id'].values[0]

# Filter runs scored by Kohli
kohli_runs = runs_df[runs_df['Striker_Id'] == kohli_player_id]

# Separate Kohli's performances when he opened and when he played at 3
kohli_opening_runs = kohli_runs[kohli_runs['Striker_Batting_Position'] == 1]
kohli_position_3_runs = kohli_runs[kohli_runs['Striker_Batting_Position'] == 3]

# Calculate the totakl runs when opening
kohli_opening_avg = kohli_opening_runs['Batsman_Scored'].sum()

# Calculate the total runs when playing at 3
kohli_position_3_avg = kohli_position_3_runs['Batsman_Scored'].sum()

# Output the average runs
print(f"Kohli's  Runs when Opening: {kohli_opening_avg:.2f}")
print(f"Kohli's  Runs when Playing at Position 3: {kohli_position_3_avg:.2f}")

# Determine if Kohli is more successful at 3
if kohli_position_3_avg > kohli_opening_avg:
    print("Kohli is more successful when playing at position 3.")
else:
    print("Kohli is more successful when opening the innings.")
```

```python
#Task-8 

raina_id=player_df[player_df['Player_Name']=='SK Raina']['Player_Id'].values[0]
ashwin_id=player_df[player_df['Player_Name']=='R Ashwin']['Player_Id'].values[0]

# Filter matches where Raina scored 30+ runs
raina_30_plus = runs_df[(runs_df['Striker_Id'] == raina_id) & (runs_df['Batsman_Scored'].sum() >= 30)]
raina_30_plus_matches = raina_30_plus['Match_Id'].unique()

# Filter matches where Ashwin took 2+ wickets
# Assuming 'dismissal_kind' indicates a wicket and excluding 'run out'
ashwin_wickets = runs_df[(runs_df['Player_dissimal_Id'] == ashwin_id) & runs_df['Dissimal_Type'].notna() & (runs_df['Dissimal_Type'] != 'run out')]
ashwin_2_plus_wickets = ashwin_wickets.groupby('Match_Id').size().reset_index(name='wickets')
ashwin_2_plus_wickets_matches = ashwin_2_plus_wickets[ashwin_2_plus_wickets['wickets'] >= 2]['Match_Id'].unique()

# Identify matches where both conditions are met
combined_matches = set(raina_30_plus_matches) & set(ashwin_2_plus_wickets_matches)

# Filter matches to include only those where CSK played
csk_matches = match_df[(match_df['Team_Name_Id'] == 'Chennai Super Kings') | (match_df['Opponent_Team_Id'] == 'Chennai Super Kings')]

# Find out how many of the combined matches were won by CSK
csk_combined_wins = csk_matches[(csk_matches['Match_Id'].isin(combined_matches)) & (csk_matches['Match_Winner_Id'] == 'Chennai Super Kings')]


# Calculate the probability
total_combined_matches = len(combined_matches)
csk_wins_count = len(csk_combined_wins)

if total_combined_matches > 0:
    probability = csk_wins_count / total_combined_matches
    print(f"The probability of CSK winning a match if Raina scores 30+ and Ashwin takes 2+ wickets is: {probability:.2%}")
else:
    print("No matches found where both Raina scored 30+ runs and Ashwin took 2+ wickets.")
```
